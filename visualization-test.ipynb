```python
import numpy as np
import glob
import matplotlib.pyplot as plt
from vis_utils import visualize_volume
from scipy.ndimage import zoom
```

```python
source_dir = './__8bit/'
files = glob.glob(source_dir + '/*/*.raw')
files[12]
```

```python
def load_volume(file, dims_order=(0, 1, 2)):
    raw_arr = np.fromfile(file, dtype=np.uint8)
    volume_dims_str = file.split('_')[-1].split('.')[0].split('x')
    volume_dims = [int(volume_dims_str[i]) for i in dims_order]
    return raw_arr.reshape(volume_dims)
```

```python
%%time
sample_volume = load_volume(files[12], dims_order=(2, 1, 0))
print(sample_volume.shape)
```

```python
visualize_volume(sample_volume, scale=0.4)
```
